```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('movies.csv')
df.shape
```

```python
df
```

```python
selected_features = ['genres' , 'keywords' , 'tagline' , 'cast' , 'director']
selected_features
```

```python
for i in selected_features :
    df[i].fillna('')
combined_features = df['genres'] + ' ' + df['keywords'] + ' ' + df['tagline'] + ' ' + df['cast'] + ' ' + df['director']
combined_features.dropna(inplace=True)
combined_features
```

```python
list_of_all_titles = df['title'].tolist()
list_of_all_titles
```

```python
movie_name = input("Enter your favorite movie name : ")
```

```python
movie_name2 = input("Enter your favorite movie name : ")
```

```python
movie_name3 = input("Enter your favorite movie name : ")
find_close_title = difflib.get_close_matches(movie_name3 , list_of_all_titles)
find_close_title
index_of_the_movie = df[df.title == find_close_title[0]]['index'].values[0]
index_of_the_movie
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
print(similarity.shape)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))
similarity_score
```

```python
sorted_similar_movies = sorted(similarity_score , key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = df[df.index == index]['title'].values[0]
    if i<10 :
        print(i , ',' , title_from_index)
        i += 1    
```
